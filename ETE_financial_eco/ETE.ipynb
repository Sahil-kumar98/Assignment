# Q2  calculate all possible correlations, covariances and variances of returns for INFOSYS, ITC and NIFTY

```python
## Finding mean of given data
def mean_value(data) :
    return np.mean(data)

## finding standard deviation of data
def std_dev(data) :
    mean_data_sqr = mean_value(data**2)
    sqr_mean_value = mean_value(data)**2
    return np.sqrt(mean_data_sqr-sqr_mean_value)

## finding variance of data
def variance_data(data) :
    mean_data_sqr = mean_value(data**2)
    sqr_mean_value = mean_value(data)*mean_value(data)
    return np.sqrt(mean_data_sqr-sqr_mean_value)

## finding coveriance between given data1 and data2
def cov_variance_data(data1, data2) :
    cov_variance = mean_value(data1*data2)- mean_value(data1)*mean_value(data2)
    return   cov_variance

## finding correlation between given data1 and data2
def correlation(data1, data2) :
    cov_variance = mean_value(data1*data2)- mean_value(data1)*mean_value(data2)
    
    return   cov_variance/(std_dev(data1)*std_dev(data2))
```

```python
## finding Variance 
print('vatinace of rturn for ITC',variance_data(ITC_daily_rturn) )
print('vatinace of rturn for INFOSYS',variance_data(INFOSYS_daily_rturn) )
print('vatinace of rturn for NIFTY',variance_data(NIFTY_daily_rturn) )
```

```python
## Finding correlation 
print('correlation between ITC and infosis', correlation(ITC_daily_rturn,INFOSYS_daily_rturn ))
print('correlation  between ITC and Nifty', correlation(ITC_daily_rturn,NIFTY_daily_rturn))
print('correlation between  infosis and Nifty', correlation(INFOSYS_daily_rturn,NIFTY_daily_rturn))
```

```python
## finding covariance
print('covariance between ITC and infosis',cov_variance_data(ITC_daily_rturn,INFOSYS_daily_rturn ))
print('covariance  between ITC and Nifty',cov_variance_data(ITC_daily_rturn,NIFTY_daily_rturn))
print('covariance between  infosis and Nifty',cov_variance_data(INFOSYS_daily_rturn,NIFTY_daily_rturn))
```
